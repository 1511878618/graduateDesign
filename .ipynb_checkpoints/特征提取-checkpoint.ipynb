```python
BCEs = pd.read_csv('milk allergy linear BCEs bovine dataset.csv')
nonBCEs = pd.read_csv('milk allergy linear non-BCEs experiece confirmed.csv')
BCEs.head(5)
```

```python
#保存序列
BCEsseq = BCEs['Description'].apply(lambda x : Seq.Seq(x))
nonBCEsseq = nonBCEs ['Description'].apply(lambda x : Seq.Seq(x))
BCEsseq.index = BCEs['Epitope ID']
nonBCEsseq.index = nonBCEs['Epitope ID']
```

```python
#把Epitope ID信息添加到Seq 里面
BCEsrecords =[]
for ID,seq in BCEsseq.items():
    BCEsrecords.append(SeqRecord.SeqRecord(seq,id=str(ID),description=''))
filename = 'milk allergy linear BCEs bovine dataset.fasta'
SeqIO.write(BCEsrecords,filename,'fasta')

nonBCEsrecords =[]
for ID,seq in nonBCEsseq.items():
    nonBCEsrecords.append(SeqRecord.SeqRecord(seq,id=str(ID),description=''))
filename = 'milk allergy linear nonBCEs bovine dataset.fasta'
SeqIO.write(nonBCEsrecords,filename,'fasta')
```

```python
nonBCEsrecords =[]
for ID,seq in nonBCEsseq.items():
    nonBCEsrecords.append(SeqRecord.SeqRecord(seq,id=str(ID),description=''))
filename = 'milk allergy linear nonBCEs bovine dataset.fasta'
SeqIO.write(nonBCEsrecords,filename,'fasta')
```

```python
def countFeatures(data):
    import feature 
    """
    input: 输入BCEs或者非BCEs的数据集 pd.DataFrame类型
    output:输出包含AAC、DPC、CTD和AAI的 pd.DataFrame类型
    
    """
    #计算AAC
    data['AAC']= data['Description'].apply(fe.CalculateAAComposition)
    #计算DPC
    data['DPC']= data['Description'].apply(fe.CalculateDipeptideComposition)
    #计算CTD
    data['CTD']= data['Description'].apply(fe.CalculateCTD)
    #计算AAI
    data['AAI']= data['Description'].apply(fe.CalculateAAIndex)
    return data
```

```python
BCEs = countFeatures(BCEs)
nonBCEs = countFeatures(nonBCEs)
```

```python
def drawfig(data,feature,draw75=True,filename=''):
    
    
    """
    适合于featureData 和AAC的处理
    input:feature 是一个 pd.DataFrame
    
    """   
    import seaborn as sns 
    from IPython.core.display import display, HTML
    import matplotlib.pyplot as plt 
    %matplotlib inline 
    %config InlineBackend.figure_format = 'jpg' #以矢量图的形式显示图片，这样比较清晰
    plt.style.use({'figure.figsize':(20,16)})
    
    featureData = data[feature]   
    #获取featureData的名字表
    for i in iter(featureData):
        featureDataNames = list(i.keys())
        break 
    #从1个Seriese中提取所有数据，Series中的每一个数据都是一个字典。
    featureData_list = [np.array([value for key,value in values.items()]) for ID,values in featureData.items()]

  
    featureData_array = sum(featureData_list)/len(featureData_list)#对每一个特征的的频率求均值
    
    featureDataData = pd.DataFrame(featureData_array.tolist(),index=featureDataNames,columns=['values'])
    #获取频率最高的前25的数据
    featureDataData75 = featureDataData[featureDataData['values'] > float(featureDataData.describe().iloc[6])]
    print("{} 的频率前10是（大到小的顺序）".format(feature)+'、'.join(list(featureDataData.sort_values(by='values',ascending=False).iloc[0:10].index)))
    
    if draw75:
        x = featureDataData75.index
        y = featureDataData75.values.flatten()
        fig = sns.barplot(x=x,y=y,linewidth= 30)
        plt.xticks(fontsize=5)
        #X坐标是以0,1,2,的顺序排布，Y坐标则是数值，name是文字
        for name,y,index in zip(x,y,range(len(x))):
                #在柱状图上绘制该类别的数量
                fig.text(index,y,name,color="blue",ha="center",fontsize=14)
        scatter_fig = fig.get_figure()
        scatter_fig.savefig(filename+feature+'75.jpg', dpi = 500)
        
    else:
        x = featureDataData.index
        y = featureDataData.values.flatten()
        fig = sns.barplot(x=x,y=y,linewidth= 30)
        plt.xticks(fontsize=5)
        #X坐标是以0,1,2,的顺序排布，Y坐标则是数值，name是文字
        for name,y,index in zip(x,y,range(len(x))):
                #在柱状图上绘制该类别的数量
                fig.text(index,y,name,color="blue",ha="center",fontsize=14)

        scatter_fig = fig.get_figure()
        scatter_fig.savefig(filename+feature+'.jpg', dpi = 500)
```

```python
drawfig(BCEs,'AAC',draw75=False,filename='BCEs')
drawfig(BCEs,'DPC',draw75=False,filename='BCEs') 
drawfig(BCEs,'DPC',draw75=True,filename='BCEs') 


drawfig(nonBCEs,'AAC',draw75=False,filename='nonBCEs')
drawfig(nonBCEs,'DPC',draw75=False,filename='nonBCEs')
drawfig(nonBCEs,'DPC',draw75=True,filename='nonBCEs')
```
